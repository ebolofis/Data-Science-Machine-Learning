```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.covariance import EllipticEnvelope
from sklearn.decomposition import PCA
import numpy as np
import requests
from io import StringIO

# URL to import the dataset from GitHub
url = 'https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/engine.csv'

def load_data_from_url(url):
    """Loads data from a given URL."""
    try:
        response = requests.get(url)
        response.raise_for_status()
        csv_content = StringIO(response.text)
        df = pd.read_csv(csv_content)

        # Check for missing values
        print("Missing values per column:")
        print(df.isnull().sum())

        # Check for duplicate rows
        print("\nNumber of duplicate rows:")
        print(df.duplicated().sum())

        return df
    except requests.exceptions.RequestException as e:
        print(f"Error loading data from URL: {e}")
        return None
    except pd.errors.ParserError as e:
        print(f"Error parsing CSV data: {e}")
        return None

def visualize_engine_data(url):
    """
    Visualizes the distribution of each feature using histograms and box plots.
    """
    df = load_data_from_url(url)
    if df is None:
        return None

    required_columns = [
        'Engine rpm', 'Lub oil pressure', 'Fuel pressure',
        'Coolant pressure', 'lub oil temp', 'Coolant temp'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return None

    for col in required_columns:
        plt.figure(figsize=(12, 6))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')

        # Box Plot
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f'Box Plot of {col}')

        plt.tight_layout()
        plt.show()

    return df  # Return the dataframe


def analyze_engine_data(df):
    """Calculates basic statistics."""
    if df is None:
        return

    required_columns = [
        'Engine rpm', 'Lub oil pressure', 'Fuel pressure',
        'Coolant pressure', 'lub oil temp', 'Coolant temp'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return

    statistics = df[required_columns].describe()
    medians = df[required_columns].median()
    statistics.loc['median'] = medians

    # Calculate 95th percentile
    percentile_95 = df[required_columns].quantile(0.95)
    print("\n95th Percentile for each feature:")
    print(percentile_95)

    statistics.loc['95th percentile'] = percentile_95  # Add 95th percentile to stats

    print("\nDescriptive Statistics with Median and 95th Percentile:")
    print(statistics)


def feature_engineering_and_analysis(df):
    """Performs feature engineering, correlation analysis, and scaling."""
    if df is None:
        return

    df.rename(columns={
        'Lub oil pressure': 'Lubrication oil pressure',
        'lub oil temp': 'Lubrication oil temperature',
        'Coolant temp': 'Coolant temperature'
    }, inplace=True)

    required_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return

    df['coolant_temp_pressure_ratio'] = df['Coolant temperature'] / df['Coolant pressure']
    df['lub_oil_temp_pressure_ratio'] = df['Lubrication oil temperature'] / df['Lubrication oil pressure']

    numerical_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature',
        'coolant_temp_pressure_ratio', 'lub_oil_temp_pressure_ratio'
    ]

    correlation_matrix = df[numerical_columns].corr()

    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

    scaler = StandardScaler()
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

    print("Feature scaling complete.")


def anomaly_detection(df, contamination=0.01, n_components=4):
    """Performs anomaly detection using various techniques and plots results with separate outlier plots, including PCA."""
    if df is None:
        return

    df.rename(columns={
        'Lub oil pressure': 'Lubrication oil pressure',
        'lub oil temp': 'Lubrication oil temperature',
        'Coolant temp': 'Coolant temperature'
    }, inplace=True)

    numerical_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature'
    ]

    df['coolant_temp_pressure_ratio'] = df['Coolant temperature'] / df['Coolant pressure']
    df['lub_oil_temp_pressure_ratio'] = df['Lubrication oil temperature'] / df['Lubrication oil pressure']

    numerical_columns.extend(['coolant_temp_pressure_ratio', 'lub_oil_temp_pressure_ratio'])

    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[numerical_columns])
    scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns)

    # Apply PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(scaled_df)
    principal_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])

    # 1. IQR Anomaly Detection
    print("\nIQR Anomaly Detection (PCA):")
    outlier_columns_iqr = []  # To store outlier columns

    for col in principal_df.columns:
        Q1 = principal_df[col].quantile(0.25)
        Q3 = principal_df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers_iqr = (principal_df[col] < (Q1 - 1.5 * IQR)) | (principal_df[col] > (Q3 + 1.5 * IQR))
        num_outliers = outliers_iqr.sum()
        print(f"Number of IQR outliers in {col}: {num_outliers}")
        outlier_columns_iqr.append(col)  # Keep track of outlier columns

        # Create binary outlier column
        principal_df[col + '_outlier_iqr'] = outliers_iqr.astype(int)

        plt.figure(figsize=(10, 6))
        normal_data = principal_df[~outliers_iqr]
        outlier_data = principal_df[outliers_iqr]
        plt.scatter(normal_data.index, normal_data[col], c='blue', s=5, label='Normal Data')
        plt.scatter(outlier_data.index, outlier_data[col], c='orange', s=15, label='Outliers')
        plt.title(f'IQR Anomaly Detection (PCA) - {col}')
        plt.legend()
        plt.show()

    # Combined IQR Outlier Calculation
    principal_df['combined_outlier_iqr'] = 0
    threshold = 2  # Adjust this threshold

    for index, row in principal_df.iterrows():
        outlier_count = 0
        for col in outlier_columns_iqr:
            outlier_count += row[col + '_outlier_iqr']
        if outlier_count >= threshold:
            principal_df.loc[index, 'combined_outlier_iqr'] = 1

    combined_outlier_percentage = (principal_df['combined_outlier_iqr'].sum() / len(principal_df)) * 100
    print(f"\nCombined IQR Outliers (Threshold = {threshold}): {combined_outlier_percentage:.2f}%")

    # IQR Effectiveness Discussion
    print("\nIQR Effectiveness Discussion:")
    print("The IQR method identifies outliers based on the spread of the data. Features with a larger IQR tend to have more data points classified as outliers.")
    print("The choice of the threshold significantly impacts the number of combined outliers. A higher threshold reduces the number of data points classified as combined outliers.")

    # 2. Z-score Anomaly Detection
    print("\nZ-score Anomaly Detection (PCA):")
    for col in principal_df.columns:
        if '_outlier_iqr' not in col and 'combined_outlier_iqr' not in col:  # Avoid Z-score on outlier flags
            z_scores = np.abs((principal_df[col] - principal_df[col].mean()) / principal_df[col].std())
            outliers_zscore = z_scores > 3
            print(f"Number of Z-score outliers in {col}: {outliers_zscore.sum()}")

            plt.figure(figsize=(10, 6))
            normal_data = principal_df[~outliers_zscore]
            outlier_data = principal_df[outliers_zscore]
            plt.scatter(normal_data.index, normal_data[col], c='blue', s=5, label='Normal Data')
            plt.scatter(outlier_data.index, outlier_data[col], c='orange', s=15, label='Outliers')
            plt.title(f'Z-score Anomaly Detection (PCA) - {col}')
            plt.legend()
            plt.show()

    # 3. Multivariate Gaussian (Elliptic Envelope)
    print("\nMultivariate Gaussian Anomaly Detection (PCA):")
    envelope = EllipticEnvelope(contamination=contamination)
    outliers_gaussian = envelope.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1  # Exclude combined outlier column
    print(f"Number of Gaussian outliers: {outliers_gaussian.sum()}")

    for col in principal_df.columns:
        if 'combined_outlier_iqr' not in col:
            plt.figure(figsize=(10, 6))
            normal_data = principal_df[envelope.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == 1]
            outlier_data = principal_df[envelope.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1]
            plt.scatter(normal_data.index, normal_data[col], c='blue', s=5, label='Normal Data')
            plt.scatter(outlier_data.index, outlier_data[col], c='orange', s=15, label='Outliers')
            plt.title(f'Multivariate Gaussian Anomaly Detection (PCA) - {col}')
            plt.legend()
            plt.show()

    # 4. Isolation Forest
    print("\nIsolation Forest Anomaly Detection (PCA):")
    iso_forest = IsolationForest(contamination=contamination)
    outliers_iso = iso_forest.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1  # Exclude combined outlier column
    num_outliers_iso = outliers_iso.sum()
    print(f"Number of Isolation Forest outliers: {num_outliers_iso}")

    for col in principal_df.columns:
        if 'combined_outlier_iqr' not in col:
            plt.figure(figsize=(10, 6))
            normal_data = principal_df[iso_forest.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == 1]
            outlier_data = principal_df[iso_forest.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1]
            plt.scatter(normal_data.index, normal_data[col], c='blue', s=5, label='Normal Data')
            plt.scatter(outlier_data.index, outlier_data[col], c='orange', s=15, label='Outliers')
            plt.title(f'Isolation Forest Anomaly Detection (PCA) - {col}')
            plt.legend()
            plt.show()

    # 5. One-Class SVM
    print("\nOne-Class SVM Anomaly Detection (PCA):")
    one_class_svm = OneClassSVM(nu=contamination)
    outliers_svm = one_class_svm.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1  # Exclude combined outlier column
    num_outliers_svm = outliers_svm.sum()
    print(f"Number of One-Class SVM outliers: {num_outliers_svm}")

    for col in principal_df.columns:
        if 'combined_outlier_iqr' not in col:
            plt.figure(figsize=(10, 6))
            normal_data = principal_df[one_class_svm.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == 1]
            outlier_data = principal_df[one_class_svm.fit_predict(principal_df.drop(columns=['combined_outlier_iqr'])) == -1]
            plt.scatter(normal_data.index, normal_data[col], c='blue', s=5, label='Normal Data')
            plt.scatter(outlier_data.index, outlier_data[col], c='orange', s=15, label='Outliers')
            plt.title(f'One-Class SVM Anomaly Detection (PCA) - {col}')
            plt.legend()
            plt.show()

    # SVM and Isolation Forest Effectiveness Discussion
    print("\nSVM and Isolation Forest Effectiveness Discussion:")
    print(f"Isolation Forest detected {num_outliers_iso} outliers, while One-Class SVM detected {num_outliers_svm} outliers.")
    print("Both methods identify anomalies without requiring labeled data, but their sensitivity depends on parameters like 'contamination'.")
    print("Isolation Forest tends to isolate anomalies more effectively when they are few and different, while SVM tries to define a boundary around normal data, which can be less effective if anomalies are close to normal instances.")

# Call the functions
df = visualize_engine_data(url)
if df is not None :
    analyze_engine_data(df)
    feature_engineering_and_analysis(df)
    anomaly_detection(df)
```
