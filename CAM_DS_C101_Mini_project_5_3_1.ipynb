As you work through this project, keep in mind that, typically speaking, anomalies would make up a minority of the data points (i.e., about 1% to 5% of the data points would be anomalies).

The data set contains six important features continuously monitored to evaluate the engine's status as ‘good’ or ‘bad’. These features are:
- **Engine rpm (revolutions per minute):** A high rpm indicates the engine is operating at a higher speed than designed for prolonged periods, which can lead to overheating, excessive wear, and eventual failure. A low rpm could signal a lack of power, issues with fuel delivery, or internal mechanical problems.
- **Lubrication oil pressure:** Low lubrication oil pressure indicates insufficient lubrication, leading to increased friction, overheating, and engine damage. A high lubrication oil pressure could signal a blockage in the oil delivery system, potentially causing seal or gasket failure.
- **Fuel pressure:** High fuel pressure can cause poor engine performance and incomplete combustion, indicating fuel pump or filter issues. A low fuel pressure may result in excessive fuel consumption, poor emissions, or damage to the fuel injectors.
- **Coolant pressure:** Low coolant pressure indicates a potential leak in the cooling system or a coolant pump failure, risking engine overheating. A high coolant pressure could be a sign of a blockage in the cooling system or a failing head gasket, which can also lead to overheating.
- **Lubrication oil temperature:** High lubrication oil temperature suggests the oil is overheating, which can degrade its lubricating properties and lead to engine damage. A low lubrication oil temperature may indicate it is not reaching its optimal operating temperature, potentially causing inadequate lubrication.
- **Coolant temperature:** High coolant temperature signals overheating, which various issues, including a failed thermostat, coolant leak, or insufficient coolant flow can cause. A low coolant temperature could suggest the engine is not reaching its optimal operating temperature, affecting performance and efficiency.

Issues with engines could lead to engine malfunctions, potential safety hazards, and downtime (e.g. delayed deliveries), resulting in the breakdown of a ship’s overall functionality, consequently impacting the business, such as affecting revenue via failure to deliver goods. By predicting timely maintenance, the business aims to increase profit by reducing downtime, reducing safety risks for the crew, limiting fuel consumption, and increasing customer satisfaction through timely deliveries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.covariance import EllipticEnvelope
import numpy as np
import requests
from io import StringIO

# URL to import the dataset from GitHub
url = 'https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/engine.csv'

def load_data_from_url(url):
    """Loads data from a given URL."""
    try:
        response = requests.get(url)
        response.raise_for_status()
        csv_content = StringIO(response.text)
        df = pd.read_csv(csv_content)
        #print("Columns in DataFrame:", df.columns)
        return df
    except requests.exceptions.RequestException as e:
        print(f"Error loading data from URL: {e}")
        return None
    except pd.errors.ParserError as e:
        print(f"Error parsing CSV data: {e}")
        return None

# PHASE 1 DATA EXPLORATION AND PREPROCESSING

def visualize_engine_data(url):
    """
    Visualizes the distribution of each feature using histograms and box plots.
    """
    df = load_data_from_url(url)
    if df is None:
        return None

    required_columns = [
        'Engine rpm', 'Lub oil pressure', 'Fuel pressure',
        'Coolant pressure', 'lub oil temp', 'Coolant temp'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return None

    for col in required_columns:
        plt.figure(figsize=(12, 6))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')

        # Box Plot
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f'Box Plot of {col}')

        plt.tight_layout()
        plt.show()

    return df  # Return the dataframe


def analyze_engine_data(df):
    """Calculates basic statistics."""
    if df is None:
        return

    required_columns = [
        'Engine rpm', 'Lub oil pressure', 'Fuel pressure',
        'Coolant pressure', 'lub oil temp', 'Coolant temp'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return

    statistics = df[required_columns].describe()
    medians = df[required_columns].median()
    statistics.loc['median'] = medians
    print(statistics)


def feature_engineering_and_analysis(df):
    """Performs feature engineering, correlation analysis, and scaling."""
    if df is None:
        return

    df.rename(columns={
        'Lub oil pressure': 'Lubrication oil pressure',
        'lub oil temp': 'Lubrication oil temperature',
        'Coolant temp': 'Coolant temperature'
    }, inplace=True)

    required_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature'
    ]

    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: Missing required columns: {missing_cols}")
        return

    df['coolant_temp_pressure_ratio'] = df['Coolant temperature'] / df['Coolant pressure']
    df['lub_oil_temp_pressure_ratio'] = df['Lubrication oil temperature'] / df['Lubrication oil pressure']

    numerical_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature',
        'coolant_temp_pressure_ratio', 'lub_oil_temp_pressure_ratio'
    ]

    correlation_matrix = df[numerical_columns].corr()

    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

    scaler = StandardScaler()
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

    print("Feature scaling complete.")


# END OF PHASE 1 DATA EXPLORATION AND PREPROCESSING

def anomaly_detection(df):
    """Performs anomaly detection using various techniques and plots results with orange outliers."""
    if df is None:
        return

    df.rename(columns={
        'Lub oil pressure': 'Lubrication oil pressure',
        'lub oil temp': 'Lubrication oil temperature',
        'Coolant temp': 'Coolant temperature'
    }, inplace=True)

    numerical_columns = [
        'Engine rpm', 'Lubrication oil pressure', 'Fuel pressure',
        'Coolant pressure', 'Lubrication oil temperature', 'Coolant temperature'
    ]

    df['coolant_temp_pressure_ratio'] = df['Coolant temperature'] / df['Coolant pressure']
    df['lub_oil_temp_pressure_ratio'] = df['Lubrication oil temperature'] / df['Lubrication oil pressure']

    numerical_columns.extend(['coolant_temp_pressure_ratio', 'lub_oil_temp_pressure_ratio'])

    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[numerical_columns])
    scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns)

    # 1. IQR Anomaly Detection
    print("\nIQR Anomaly Detection:")
    outliers_iqr = pd.Series(index=scaled_df.index, dtype=bool)
    for col in numerical_columns:
        Q1 = scaled_df[col].quantile(0.25)
        Q3 = scaled_df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers_iqr = outliers_iqr | ((scaled_df[col] < (Q1 - 1.5 * IQR)) | (scaled_df[col] > (Q3 + 1.5 * IQR)))
    print(f"Number of IQR outliers: {outliers_iqr.sum()}")

    plt.figure(figsize=(10, 6))
    for col in numerical_columns:
        colors = ['orange' if outlier else 'blue' for outlier in outliers_iqr]
        plt.scatter(scaled_df.index, scaled_df[col], c=colors, s=5, label=col)
    plt.title('IQR Anomaly Detection')
    plt.legend()
    plt.show()

    # 2. Z-score Anomaly Detection
    print("\nZ-score Anomaly Detection:")
    outliers_zscore = pd.Series(index=scaled_df.index, dtype=bool)
    for col in numerical_columns:
        z_scores = np.abs((scaled_df[col] - scaled_df[col].mean()) / scaled_df[col].std())
        outliers_zscore = outliers_zscore | (z_scores > 3)
    print(f"Number of Z-score outliers: {outliers_zscore.sum()}")

    plt.figure(figsize=(10, 6))
    for col in numerical_columns:
        colors = ['orange' if outlier else 'blue' for outlier in outliers_zscore]
        plt.scatter(scaled_df.index, scaled_df[col], c=colors, s=5, label=col)
    plt.title('Z-score Anomaly Detection')
    plt.legend()
    plt.show()

    # 3. Multivariate Gaussian (Elliptic Envelope)
    print("\nMultivariate Gaussian Anomaly Detection:")
    envelope = EllipticEnvelope(contamination=0.05)
    outliers_gaussian = envelope.fit_predict(scaled_df) == -1
    print(f"Number of Gaussian outliers: {outliers_gaussian.sum()}")
    plt.figure(figsize=(10, 6))
    for col in numerical_columns:
        colors = ['orange' if outlier else 'blue' for outlier in outliers_gaussian == -1]
        plt.scatter(scaled_df.index, scaled_df[col], c=colors, s=5, label=col)
    plt.title('Multivariate Gaussian Anomaly Detection')
    plt.legend()
    plt.show()

    # 4. Isolation Forest
    print("\nIsolation Forest Anomaly Detection:")
    iso_forest = IsolationForest(contamination=0.05)
    outliers_iso = iso_forest.fit_predict(scaled_df) == -1
    print(f"Number of Isolation Forest outliers: {outliers_iso.sum()}")
    plt.figure(figsize=(10, 6))
    for col in numerical_columns:
        colors = ['orange' if outlier else 'blue' for outlier in outliers_iso == -1]
        plt.scatter(scaled_df.index, scaled_df[col], c=colors, s=5, label=col)
    plt.title('Isolation Forest Anomaly Detection')
    plt.legend()
    plt.show()

    # 5. One-Class SVM
    print("\nOne-Class SVM Anomaly Detection:")
    one_class_svm = OneClassSVM(nu=0.05)
    outliers_svm = one_class_svm.fit_predict(scaled_df) == -1
    print(f"Number of One-Class SVM outliers: {outliers_svm.sum()}")
    plt.figure(figsize=(10, 6))
    for col in numerical_columns:
        colors = ['orange' if outlier else 'blue' for outlier in outliers_svm == -1]
        plt.scatter(scaled_df.index, scaled_df[col], c=colors, s=5, label=col)
    plt.title('One-Class SVM Anomaly Detection')
    plt.legend()
    plt.show()

# Call the functions
df = visualize_engine_data(url)
if df is not None :
    analyze_engine_data(df)
    feature_engineering_and_analysis(df)
    anomaly_detection(df)
```
